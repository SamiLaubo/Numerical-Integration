##### Hénon-Heiles modell og hjelpefunksjoner
Funksjonene *F* og *F_scipy* implementerer Hénon-Heiles modell for hhv. de numeriske metodene og scipy.integrate.solve_ivp. Energi i form av potenseill energi og Hamiltonfunksjonen er i funksjonene *U* og *H*. Interpolasjonsfunksjonen interpolerer $q_2$ og $p_2$ til punktet med tilhørende $q_1$-verdi lik 0. I tillegg er en egen modulo-funksjon implementert fordi python-versjonen medførte maskinavrundningsfeil.

##### Numeriske metoder 
De fire numeriske metodene har en funksjon hver. Disse beregner y ved å lagre forrige og neste y-verdi og iterere gjennom tid. Argumentet "version" gir den siste y-verdien ("Last value"), energien for hvert hele tidssteg ("Energy") eller Poincaré-kutt ("Poincare") for metoden.

##### Parallellisering  
concurrent.futures.ProcessPoolExecutor blir brukt til å utnytte flere CPU-enheter for å kjøre beregningene for ulike metoder og steglengder parallellt.

## 3. Hénon-Heiles modell og hjelpefunksjoner

##### Hénon-Heiles modell

```python
@njit
def F(
    y,     # (np.ndarray)  - Vector with y-values
    F_vec  # (np.ndarray)  - Destination vector for computation
):
    F_vec[0] = y[2]                                         # q_1 derivert
    F_vec[1] = y[3]                                         # q_2 derivert
    F_vec[2] = -y[0] * (1 + 2 * y[1])                       # p_1 derivert
    F_vec[3] = np.power(y[1], 2) - np.power(y[0], 2) - y[1] # p_2 derivert
```

##### Hénon-Heiles modell for integrering med scipy

```python
@njit
def F_scipy(
    t,  # (float)       - Time to evaluate y, y(t)
    y   # (np.ndarray)  - Solution for previous step
):
    return np.array([
        y[2],                                         # q_1 derivative
        y[3],                                         # q_2 derivative
        -y[0] * (1 + 2 * y[1]),                       # p_1 derivative
        np.power(y[1], 2) - np.power(y[0], 2) - y[1]  # p_2 derivative
    ])
```

##### Potensiell Energi

```python
@njit
def U(
    q1, # (float)  - 1. coordinate
    q2  # (float)  - 2. coordinate
):
    return .5 * (np.power(q1, 2) + np.power(q2, 2)) + np.power(q1, 2) * q2 - 1 / 3 * np.power(q2, 3)
```

##### Hamiltonsk energi

```python
@njit
def H(
    y  # (np.ndarray)  - Vector = (q1, q2, p1, p2)
):
    return .5 * (np.power(y[2], 2) + np.power(y[3], 2)) + np.power(y[0], 2) * (y[1] + .5) + np.power(y[1], 2) * (.5 - y[1] / 3)
```

##### Interpolasjon

```python
@njit
def interpolate(
    dest,  # (np.ndarray)  - Vector for solution
    y1,    # (float)       - 1. coordinate
    y2     # (float)       - 2. coordinate
):
    lam = y1[0] / (y1[0] - y2[0])  # Lambda value for points closest to q1=0
    dest[0] = lam * y2[1] + (1 - lam) * y1[1]
    dest[1] = lam * y2[3] + (1 - lam) * y1[3]
```

##### Modulo

```python
@njit
def mod(
    a,  # (float/int)
    b   # (float/int)
):
    return a - np.floor_divide(a, b) * b
```

##### Startverdi for y

```python
y_0 = np.array([0, .45, np.sqrt((1/12 - U(0, .45)) * 2 ), 0])
```

## 4. Numeriske metoder

### 4.1. Metode A: Bogacki-Shampine av orden 3

Bogacki-Shampine av orden 3 er her definert for vilkårlige differensialligninger av formen $y'=F(y)$ med konstant steglengde $h$. $y_n$ er gjeldende verdi for løsningen og brukes til å beregne $y_{n+1}$ som er neste verdi, en steglengde lenger.  

**Bogacki-Shampine:**  
  
$F_1 = F(y_n)$  
$F_2 = F(y_n + \frac{1}{2} h F_1)$  
$F_3 = F(y_n + \frac{3}{4} h F_2)$  
$y_{n+1} = y_n + h \left(\frac{2}{9} F_1 + \frac{1}{3} F_2 + \frac{4}{9} F_3 \right)$

```python
@njit
def bogacki_shampine(
    y_0,                  # (np.ndarray)  - Initial values for y
    h,                    # (int)         - Step length
    t_0,                  # (int)         - Start time
    T,                    # (int)         - End time
    version='Last value'  # (string)      - What version to compute ('last value', 'Energy', 'Poincare')
):
    # Array to store previous y-values and initiate to y_0
    y_prev = np.empty(len(y_0))
    y_prev = np.copy(y_0)
    
    # Array to store next y-values
    y_next = np.empty(len(y_0))
    
    # Matrix to store F1, F2 and F3
    F_vec = np.zeros((3, len(y_0)))
    
    # Function that computes next step
    def step():
        # Use outer scope to enable assignment of new values ("pass by reference")
        nonlocal y_prev, y_next, F_vec, h
        
        # Solve F1, F2 and F3
        F(y_prev, F_vec[0])
        F(y_prev + .5 * h * F_vec[0], F_vec[1])
        F(y_prev + .75 * h * F_vec[1], F_vec[2])

        # Compute next y-value
        y_next = y_prev + h * (2 / 9 * F_vec[0] + 1 / 3 * F_vec[1] + 4 / 9 * F_vec[2])

    ### Last value
    if version == 'Last value':
        # Time development
        for i in range(np.int(np.floor(T / h))):
            # Compute next step
            step()
            
            # next y-value will be previous y-value in next iteration
            y_prev = np.copy(y_next)
        
        # Take one extra step to reach exactly T
        h = mod(T, h)  # Modulo operator
        step()
        
        # Return y(T)
        return y_next
    
    ### Energy
    if version == 'Energy':
        # Array to store Energy 
        H_values = np.zeros(T + 1)
        H_values[0] = H(y_0)

        # Steps to skip when storing energy, store one per whole timestep
        skip = int(1 // h)
        
        # Time development
        for i in range(T):
            # Go to next whole timestep
            for j in range(skip):
                # Compute next step
                step()
                
                # next y-value will be previous y-value in next iteration
                y_prev = np.copy(y_next)
                
            # Compute and store energy
            H_values[i + 1] = H(y_next)
        
        # Return energy values
        return H_values
    
    ### Poincaré
    if version == 'Poincare':
        # Array to store interpolation of q2 and p2
        q_p_interpolate = np.zeros(2)

        # Array to store coresponding q2, p2 pairs [q21, p21, q22, p22, ...]
        q_p_2 = np.empty(0)
        
        # Time development
        for i in range(np.int(np.floor(T / h))):
            # Compute next step
            step()

            # If previous and next q1-values have different signs and p1 is positive
            if y_prev[0] * y_next[0] < 0 and y_next[2] > 0:
                # Interpolate to find more exact value
                interpolate(q_p_interpolate, y_prev, y_next)

                # Store point
                q_p_2 = np.append(q_p_2, q_p_interpolate)

            # next y-value will be previous y-value in next iteration
            y_prev = np.copy(y_next)
        
        # Return poincaré cut points
        return q_p_2
```

### 4.2. Metode B: Kuttas metode av orden 4

Runge Kutta av orden 4 er her definert for vilkårlige differensialligninger av formen $y'=F(y)$ med konstant steglengde $h$.

**Runge Kutta:**  
  
$F_1 = F(y_n)$  
$F_2 = F(y_n + \frac{1}{2} h F_1)$  
$F_3 = F(y_n + \frac{1}{2} h F_2)$  
$F_4 = F(y_n + h F_3)$  
$y_{n+1} = y_n + h \left(\frac{1}{6} F_1 + \frac{1}{3} F_2 + \frac{1}{3} F_3  + \frac{1}{6} F_4 \right)$

```python
@njit
def kutta(
    y_0,                  # (np.ndarray)  - Initial values for y
    h,                    # (int)         - Step length
    t_0,                  # (int)         - Start time
    T,                    # (int)         - End time
    version='Last value'  # (string)      - What version to compute ('last value', 'Energy', 'Poincare')
):
    # Array to store previous y-values and initiate to y_0
    y_prev = np.empty(len(y_0))
    y_prev = np.copy(y_0)
    
    # Array to store next y-values
    y_next = np.empty(len(y_0))
    
    # Matrix to store F1, F2 and F3
    F_vec = np.zeros((4, len(y_0)))
    
    # Function that computes next step
    def step():
        # Use outer scope to enable assignment of new values ("pass by reference")
        nonlocal y_prev, y_next, F_vec, h
        
        # Solve F1, F2, F3 and F4
        F(y_prev, F_vec[0])
        F(y_prev + .5 * h * F_vec[0], F_vec[1])
        F(y_prev + .5 * h * F_vec[1], F_vec[2])
        F(y_prev + h * F_vec[2], F_vec[3])

        # Compute next y-value
        y_next = y_prev + h * (1 / 6 * F_vec[0] + 1 / 3 * F_vec[1] + 1 / 3 * F_vec[2] + 1 / 6 * F_vec[3])

    
    ### Last value
    if version == 'Last value':
        # Time development
        for i in range(np.int(np.floor(T / h))):
            # Compute next step
            step()
            
            # next y-value will be previous y-value in next iteration
            y_prev = np.copy(y_next)
        
        # Take one extra step to reach exactly T
        h = mod(T, h)  # Modulo operator
        step()
        
        # Return y(T)
        return y_next
    
    ### Energy
    if version == 'Energy':
        # Array to store Energy 
        H_values = np.zeros(T + 1)
        H_values[0] = H(y_0)

        # Steps to skip when storing energy, store one per whole timestep
        skip = int(1 // h)
        
        # Time development
        for i in range(T):
            # Go to next whole timestep
            for j in range(skip):
                # Compute next step
                step()
                
                # next y-value will be previous y-value in next iteration
                y_prev = np.copy(y_next)
                
            # Compute and store energy
            H_values[i + 1] = H(y_next)
        
        # Return energy values
        return H_values
    
    
    ### Poincaré
    if version == 'Poincare':
        # Array to store interpolation of q2 and p2
        q_p_interpolate = np.zeros(2)

        # Array to store coresponding q2, p2 pairs [q21, p21, q22, p22, ...]
        q_p_2 = np.empty(0)
        
        # Time development
        for i in range(np.int(np.floor(T / h))):
            # Compute next step
            step()

            # If previous and next q1-values have different signs and p1 is positive
            if y_prev[0] * y_next[0] < 0 and y_next[2] > 0:
                # Interpolate to find more exact value
                interpolate(q_p_interpolate, y_prev, y_next)

                # Store point
                q_p_2 = np.append(q_p_2, q_p_interpolate)

            # next y-value will be previous y-value in next iteration
            y_prev = np.copy(y_next)
        
        # Return poincaré cut points
        return q_p_2
```

### 4.3. Metode C: Kahans metode av orden 2

Kahans metode av orden 2 er her definert for differensialligninger med kvadratiske høyresider. Metoden gir ligningssystemet $Ay_{n+1}=b$ der $y_{n+1,1}$ gir første komponent i neste steg.

**Kahan:**  
  
$A = \left[\begin{matrix} 1 & 0 & -\frac{h}{2} & 0 \\ 0 & 1 & 0 & -\frac{h}{2} \\ h (y_{n,2} + .5) & h y_{n,1} & 1 & 0 \\ h y_{n,1} & h (.5 - y_{n,2}) & 0 & 1 \end{matrix}\right]$  
  
$b = \left[y_{n,1}+\frac{h}{2} y_{n,3}, y_{n,2}+\frac{h}{2} y_{n,4}, y_{n,3}-\frac{h}{2} y_{n,1}, y_{n,4}-\frac{h}{2} y_{n,2}\right]^T$  

```python
@njit
def kahan(
    y_0,                  # (np.ndarray)  - Initial values for y
    h,                    # (int)         - Step length
    t_0,                  # (int)         - Start time
    T,                    # (int)         - End time
    version='Last value'  # (string)      - What version to compute ('last value', 'Energy', 'Poincare')
):
    # Array to store previous y-values and initiate to y_0
    y_prev = np.empty(len(y_0))
    y_prev = np.copy(y_0)
    
    # Array to store next y-values
    y_next = np.empty(len(y_0))
    
    # A-matrix
    A = np.zeros((4,4))
    A[0] = [1,      0, -h / 2,      0]
    A[1] = [0,      1,      0, -h / 2]
    A[2] = [0,      0,      1,      0]
    A[3] = [0,      0,      0,      1]
    
    # b-vector
    b = np.empty(4)
    
    # Function that computes next step
    def step():
        # Use outer scope to enable assignment of new values ("pass by reference")
        nonlocal y_prev, y_next, h, A, b
        
        # Update A-matrix
        A[0][2] = -h / 2
        A[1][3] = -h / 2

        A[2][0] = h * (y_prev[1] + .5)
        A[2][1] = h * y_prev[0]

        A[3][0] = h * y_prev[0]
        A[3][1] = h * (.5 - y_prev[1])

        # Update b-vector
        b[0] = y_prev[0] + h / 2 * y_prev[2]
        b[1] = y_prev[1] + h / 2 * y_prev[3]
        b[2] = y_prev[2] - h / 2 * y_prev[0]
        b[3] = y_prev[3] - h / 2 * y_prev[1]

        # Solve linear system for y and store computed y-value
        y_next = np.linalg.solve(A, b)

    
    ### Last value
    if version == 'Last value':
        # Time development
        for i in range(np.int(np.floor(T / h))):
            # Compute next step
            step()
            
            # next y-value will be previous y-value in next iteration
            y_prev = np.copy(y_next)
        
        # Take one extra step to reach exactly T
        h = mod(T, h)  # Modulo operator
        step()
        
        # Return y(T)
        return y_next
    
    ### Energy
    if version == 'Energy':
        # Array to store Energy 
        H_values = np.zeros(T + 1)
        H_values[0] = H(y_0)

        # Steps to skip when storing energy, store one per whole timestep
        skip = int(1 // h)
        
        # Time development
        for i in range(T):
            # Go to next whole timestep
            for j in range(skip):
                # Compute next step
                step()
                
                # next y-value will be previous y-value in next iteration
                y_prev = np.copy(y_next)
                
            # Compute and store energy
            H_values[i + 1] = H(y_next)
        
        # Return energy values
        return H_values
    
    
    ### Poincaré
    if version == 'Poincare':
        # Array to store interpolation of q2 and p2
        q_p_interpolate = np.zeros(2)

        # Array to store coresponding q2, p2 pairs [q21, p21, q22, p22, ...]
        q_p_2 = np.empty(0)
        
        # Time development
        for i in range(np.int(np.floor(T / h))):
            # Compute next step
            step()

            # If previous and next q1-values have different signs and p1 is positive
            if y_prev[0] * y_next[0] < 0 and y_next[2] > 0:
                # Interpolate to find more exact value
                interpolate(q_p_interpolate, y_prev, y_next)

                # Store point
                q_p_2 = np.append(q_p_2, q_p_interpolate)

            # next y-value will be previous y-value in next iteration
            y_prev = np.copy(y_next)
        
        # Return poincaré cut points
        return q_p_2
```

### 4.4. Metode D: Störmer-Verlet av orden 2

Denne metoden er laget for andreordens systemer av formen $q''=f(q)$ som omskrevet til et første ordens system ($p := q'$) blir ($q' = p$, $p' = f(q)$).

**Störmer-Verlet:**  
  
$p_{n+\frac{1}{2}}=p_n+\frac{1}{2} h f(q_n)$  
$q_{n+1}=q_n+h p_{n+\frac{1}{2}}$  
$p_{n+1}=p_{n+\frac{1}{2}}+\frac{1}{2} h f(q_{n+1})$

```python
@njit
def stormer_verlet(
    y_0,                  # (np.ndarray)  - Initial values for y
    h,                    # (int)         - Step length
    t_0,                  # (int)         - Start time
    T,                    # (int)         - End time
    version='Last value'  # (string)      - What version to compute ('last value', 'Energy', 'Poincare')
):
        # Array to store previous y-values and initiate to y_0
    y_prev = np.empty(len(y_0))
    y_prev = np.copy(y_0)
    
    # Array to store next y-values
    y_next = np.empty(len(y_0))
    
    # Matrix to store F1, F2 and F3
    F_vec = np.zeros(len(y_0))
    
    # Compute first F-vector since loop reuses previous iteration result 
    F(y_prev, F_vec)
    
    # Function that computes next step
    def step():
        # Use outer scope to enable assignment of new values ("pass by reference")
        nonlocal y_prev, y_next, F_vec, h
        
        # Update p_(n + 1/2) and temporarily store in p_(n + 1)
        y_next[2] = y_prev[2] + .5 * h * F_vec[2]
        y_next[3] = y_prev[3] + .5 * h * F_vec[3]

        # Update q
        y_next[0] = y_prev[0] + h * y_next[2]
        y_next[1] = y_prev[1] + h * y_next[3]

        # Compute F-vector
        F(y_next, F_vec)

        # Update p_(n + 1) by adding to p_(n + 1/2)
        y_next[2] += .5 * h * F_vec[2]
        y_next[3] += .5 * h * F_vec[3]

    
    ### Last value
    if version == 'Last value':
        # Time development
        for i in range(np.int(np.floor(T / h))):
            # Compute next step
            step()
            
            # next y-value will be previous y-value in next iteration
            y_prev = np.copy(y_next)
        
        # Take one extra step to reach exactly T
        h = mod(T, h)  # Modulo operator
        step()
        
        # Return y(T)
        return y_next
    
    ### Energy
    if version == 'Energy':
        # Array to store Energy 
        H_values = np.zeros(T + 1)
        H_values[0] = H(y_0)

        # Steps to skip when storing energy, store one per whole timestep
        skip = int(1 // h)
        
        # Time development
        for i in range(T):
            # Go to next whole timestep
            for j in range(skip):
                # Compute next step
                step()
                
                # next y-value will be previous y-value in next iteration
                y_prev = np.copy(y_next)
                
            # Compute and store energy
            H_values[i + 1] = H(y_next)
        
        # Return energy values
        return H_values
    
    
    ### Poincaré
    if version == 'Poincare':
        # Array to store interpolation of q2 and p2
        q_p_interpolate = np.zeros(2)

        # Array to store coresponding q2, p2 pairs [q21, p21, q22, p22, ...]
        q_p_2 = np.empty(0)
        
        # Time development
        for i in range(np.int(np.floor(T / h))):
            # Compute next step
            step()

            # If previous and next q1-values have different signs and p1 is positive
            if y_prev[0] * y_next[0] < 0 and y_next[2] > 0:
                # Interpolate to find more exact value
                interpolate(q_p_interpolate, y_prev, y_next)

                # Store point
                q_p_2 = np.append(q_p_2, q_p_interpolate)

            # next y-value will be previous y-value in next iteration
            y_prev = np.copy(y_next)
        
        # Return poincaré cut points
        return q_p_2
```

## 5. Oppgave 1 - Konvergenstest

### 5.1. Beskrivelse

Metodene verifiseres ved å sammenligne dem med en referanseløsning (scipy.integrate.solve_ivp), dette gir en globalfeil som brukes til å estimere konvergensorden. Globalfeilen blir plottet mot steglengde i figur 1 for de ulike metodene og estimert konvergensorden vises i figur 2.

### 5.2. Funksjoner

##### Beregn globalfeil

```python
def convergence_order(
    r,           # (int)         - Start point for logarithmic interval
    s,           # (int)         - End point for logarithmic interval
    num_steps,   # (int)         - Number of different step-lengths to plot for
    y_0,         # (np.ndarray)  - Initial values for y
    t_0,         # (int)         - Start time
    T,           # (int)         - End time
    rtol,        # (float)       - Relative tolerance for reference solution
    atol         # (float)       - Absolute tolerance for reference solution
):
    # Array with steps to estimate error
    h_steps = np.logspace(r, s, num_steps)
    
    # Dictionary to store global errors (one for each step)
    global_error = {
        'Bogacki-Shampine': [],
        'Kutta':            [],
        'Kahan':            [],
        'Stormer-Verlet':   []
    }
    
    # Time development
    for h in h_steps:
        # Reference solution (gets last y-value)
        y_ref = integrate.solve_ivp(F_scipy, [t_0, T], y_0, max_step=h, rtol=rtol, atol=atol, t_eval=[T]).y[:,-1]
        
        # Integrate, find error using the Euclidean norm and store
        global_error['Bogacki-Shampine'].append(np.linalg.norm(bogacki_shampine(y_0, h, t_0, T) - y_ref))
        global_error['Kutta']           .append(np.linalg.norm(kutta(y_0, h, t_0, T)            - y_ref))
        global_error['Kahan']           .append(np.linalg.norm(kahan(y_0, h, t_0, T)            - y_ref))
        global_error['Stormer-Verlet']  .append(np.linalg.norm(stormer_verlet(y_0, h, t_0, T)   - y_ref))
        
    return h_steps, global_error
```

##### Plot global feil og beregnet orden p

```python
def plot_convergence_order(
    r,             # (int)         - Start point for logarithmic interval
    s,             # (int)         - End point for logarithmic interval
    num_steps,     # (int)         - Number of different step-lengths to plot for
    y_0,           # (np.ndarray)  - Initial values
    t_0,           # (int)         - Starttime
    T,             # (int)         - Endtime
    rtol = 1e-12,  # (float)       - Relative tolerance for reference solution
    atol = 1e-12   # (float)       - Absolute tolerance for reference solution
):
    # Compute global error for step-lengths
    h_steps, global_error = convergence_order(r, s, num_steps, y_0, t_0, T, rtol, atol)
    
    # Reference line to plot against
    ref_line = {
        'Bogacki-Shampine': { 'order': 3, 'K': 1, 'color': 'r' },
        'Kutta':            { 'order': 4, 'K': 1, 'color': 'g' },
        'Kahan':            { 'order': 2, 'K': 1, 'color': 'b' },
        'Stormer-Verlet':   { 'order': 2, 'K': 1, 'color': 'y' }
    }
    
    # Storage for p-values
    p = {
        'Bogacki-Shampine': np.array([]),
        'Kutta':            np.array([]),
        'Kahan':            np.array([]),
        'Stormer-Verlet':   np.array([])
    }
    
    # Plot for convergence
    fig_convergence, axes2d = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(9,6), tight_layout=True)
    
    # Plot for p-values
    fig_p, axes_p = plt.subplots(1)
    axes_p.set_title('Konvergensorden p')
    axes_p.set_xlabel('Steglengde')
    axes_p.set_ylabel('Konvergensorden p')
    
    # Do for all methods
    for idx, item in enumerate(global_error.items()):
        method, val = item
        
        ## Plot global error
        # Fix axes
        ax = axes2d[int(idx > 1) ][int(idx % 2)]
        ax.set_title(method)
        if idx > 1: ax.set_xlabel('Steglengde')
        if not idx % 2: ax.set_ylabel('Global feil')
            
        ax.loglog(h_steps, val, color=ref_line[method]['color'])
        ax.loglog(
            h_steps, ref_line[method]['K'] * h_steps ** ref_line[method]['order'], 
            '--', color=ref_line[method]['color'], label=f'Reference line of order {ref_line[method]["order"]}'
        )
        
        # Show legend
        ax.legend(loc='lower right')
        
        ## Compute and plot p-value
        p = np.log(np.array([val[1:]]) / np.array([val[:-1]])) / np.log(np.array([h_steps[1:]]) / np.array([h_steps[:-1]]))
        
        axes_p.semilogx(h_steps[1:], p[0], label=method, color=ref_line[method]['color'])
    
    # Show legends
    axes_p.legend()
    
    # Show plots
    plt.show()
```

## 6. Oppgave 2 - Energibevarelse

### 6.1. Beskrivelse

Undersøkelse av metodenes energibevarende egenskaper for Hamiltonfunksjonen i Hénon-Heiles modellen. Dette vil vise hvor gode metodene er på å bevare systemets energi over lange tidshorisonter. Eventuelt tap av energi vil vises i figurene som en drift bort fra startenergien $H$.

### 6.2. Funksjoner

##### Beregn energi

```python
def energy_conservation(
    y_0,         # (np.ndarray)  - Initial values
    t_0,         # (int)         - Starttime
    T,           # (int)         - Endtime
    h_arr        # (array)       - Array with h-values to compute (step lengths)
):
    # Storage for energies
    energy = {}
    
    # Create processpoolexecutor
    with PPE() as executor:
        # Submit all tasks to PPE "Run all method-functions for all h-values in h_arr"
        # Tasks are run on seperate CPUs in parallell
        bogacki_shampine_processes = [executor.submit(bogacki_shampine, y_0, h, t_0, T, version='Energy') for h in h_arr]
        kutta_processes = [executor.submit(kutta, y_0, h, t_0, T, version='Energy') for h in h_arr]
        kahan_processes = [executor.submit(kahan, y_0, h, t_0, T, version='Energy') for h in h_arr]
        stormer_verlet_processes = [executor.submit(stormer_verlet, y_0, h, t_0, T, version='Energy') for h in h_arr]
        
        # Wait for all tasks for complete
        for p in as_completed([*bogacki_shampine_processes, *kutta_processes, *kahan_processes, *stormer_verlet_processes]): 
            pass
        
        # Retrieve and store an array of data for each h-value
        energy = {
            'Bogacki-Shampine': [process.result() for process in bogacki_shampine_processes],
            'Kutta':            [process.result() for process in kutta_processes],
            'Kahan':            [process.result() for process in kahan_processes],
            'Stormer-Verlet':   [process.result() for process in stormer_verlet_processes],
            'Time':             [np.arange(t_0, T + 1, 1) for h in h_arr]
        }
        
    return energy
```

##### Plot energi

```python
def plot_energy_conservation(
    y_0,         # (np.ndarray)  - Initial values
    t_0,         # (int)         - Starttime
    T,           # (int)         - Endtime
    h_arr        # (array)       - Array with h-values to compute
):
    # Get energies
    energy = energy_conservation(y_0, t_0, T, h_arr)
    # Reverse order of dictionary for plotting purposes
    energy_reversed = {}
    for i in reversed([k for k in energy.keys() if k != 'Time']): energy_reversed[i] = energy[i]
        
    # Colors for plot
    colors = {
        'Bogacki-Shampine': 'y',
        'Kutta':            'r',
        'Kahan':            'g',
        'Stormer-Verlet':   'b',
    }

    # Plot for different h values
    for idx, h in enumerate(h_arr):
        # Create figure
        fig, axes2d = plt.subplots(nrows=4, ncols=4, figsize=(9,6), tight_layout=True)#, sharex=True)
        
        # Big left plot
        main_ax = plt.subplot(1, 4, (1, 3))
        main_ax.set_title(f'Energibevarelse for h = {h}')
        
        # Plot 1 value for each timestep because matplotlib crashes for big plots
        #plot_skip = np.int(1 / h)

        # Plot all methods
        for method_idx, item in enumerate(energy_reversed.items()):
            # Get method and values
            method, val = item
            
            # Stop diverging plots
            cut_off_idx = np.argmax(val[idx] < [.083, .08333, .08333328][idx])
            if (cut_off_idx):
                cut_off_val = val[idx][:cut_off_idx]
            else:
                cut_off_val = val[idx]
            
            # Do not plot time
            if method == 'Time': continue
            
            # Plot in big left plot
            if method == 'Kahan' or method == 'Stormer-Verlet':
                # Plot main graph
                main_ax.plot(energy['Time'][idx], val[idx], label=method, color=colors[method], alpha=.1)
                # Plot average line
                skip = np.int(T / 100)
                
                average = np.average(val[idx][:np.int(np.floor_divide(len(val[idx]), skip) * skip)].reshape(-1, skip), axis=1)
                main_ax.plot(energy['Time'][idx][np.floor_divide(skip, 2)::skip], average, '--', label=f'{method} average', color=colors[method])
            else:
                main_ax.plot(energy['Time'][idx][:len(cut_off_val)], cut_off_val, label=method, color=colors[method])
                
            
            # Plot in seperate right plot
            ax = plt.subplot(4, 4, method_idx * 4 + 4)
            ax.set_title(method)
            small_plot_T = 100
            ax.plot(energy['Time'][idx][:np.int(small_plot_T)], val[idx][:np.int(small_plot_T)], color=colors[method])
            
            # Only show x-axis for bottom plot
            ax.set_yticks([])
            if method_idx * 4 + 4 == 16:
                ax.set_xticks([t_0, small_plot_T])
                ax.set_xlabel('Time')
            else:
                ax.set_xticks([])
        
        # Plot energy reference line 
        main_ax.plot([t_0, T], [1 / 12] * 2, '-.', color='k', label=r'$H = \frac{1}{12}$', alpha=.5)
        
        # axis-labels
        main_ax.set_xlabel('Time')
        main_ax.set_ylabel('Energy')
        
        # Show plot
        main_ax.legend()
        plt.show()
```

### 6.4. Diskusjon

##### Energibevarende, symplektisk og drift

Bogacki-Shampine og Kutta er metodene som under passende forhold gir en realtivt horisontal energibevarings-kurve. Plottet viser at metodene starter med lik energi for systemet men har deretter en lineær drift vekk fra denne energien. Metodene er dermed ikke energibevarende og av den lineeære driften kan man konkludere med at de i heller ikke er symplektiske [2, s.367 - teorem 8.1].  

Kahan og Störmer-Verlets metode viser derimot egenskapen at de holder seg i et omegn av startenergien, og ut ifra gjennomsnittet ser vi at startenergien ikke er eksakt. Störmer-Verlet er en symplektisk metode [2, s.414] og benytter dermed en perturbert versjon av hamiltonfunksjonen. I likhet benytter Kahans metode også en modifisert hamilton-funksjon [3] og det er dette som fører til den oscillerende oppførselen om startenergien vi ser for de to metodene. Kahan er derimot ikke symplektisk [3] og som følge av oscillasjon er hverken Störmer-Verlet eller Kahan energibevarende.

##### Steglengde og integrasjonslengde

Hvor gode metodene er til å beregne energien varierer med steglengde og integrasjonlengde. Den generelle tendensen er at kortere integrasjonsintervall og steglengde gir bedre resultater. Fra plottet ser en at Bogacki-Shampine divergerer selv for den korteste steglengden når en integrerer over intervallet $[0, 10^6]$. For Kutta ser en derimot først ved steglengde 0.1 en betydelig divergens sammenlignet med de andre metodene.

Den lineære driften hos Bogacki-Shampine og Kutta nevnt over samsvarer med hva en kan forvente av feil til en ikke-symplektisk metode [2, s.367 - teorem 8.1]. Det at Bogacki-Shampine divergerer tidligere enn Kutta skyldes ordensforskjellen. Kutta er dermed en mer robust metode som gjengir en mer eksakt energi for større steglengder og for lengre integrasjonsintervall. 

Metodene Störmer-Verlet og Kahan påvirkes også av steglengden, og ut ifra plottet ser en at de oscillerer med mindre amplitude ved mindre steglengde. Lengden på intergrasjonsintervallet ser derimot ikke ut til å ha noen synlig innvirkning på plottene. Dette kan sees i sammenheng med Teorem 8.1 [2, s.367], hvor feilen til en symplektisk integrator vises å være betraktelig mindre enn den av en ikke-symeplektisk. Störmer-Verlet bevarer altså den perturberte hamiltonfunksjonen nær eksakt over lengre tid og drifter dermed i relativt liten grad.
Khahans metode ser også ut til å ha denne egenskapen, dvs. minimalt med drift fra startenergien, muligens også på grunn av sin modifiserte hamiltonfunksjon.

##### Foretrukken metode

Ut ifra resultatene kan man konkludere med at foretrukken metode avhenger av tidssteg og integrasjonslengde. Hvis en ønsker å gjengi energien til systemet så eksakt som mulig over realtiv kort tid er det tydelig at Kutta er å foretrekke, men Bogacki-Shampine gir også gode resultater over korte integrasjonstider.

Skal en derimot integrere over svært lange intervaller kan en vurdere å benytte Kahans metode eller Störmer-Verlet. De gjengir som nevnt ikke den eksakte energien, men vil etter lengre tid gjengi den beste approksimasjonen. Basert på gjennomsnittsverdi er Störmer-Verlet å foretrekke.

## 7. Oppgave 3 - Poincaré-seksjoner

### Beskrivelse

Beregning og plotting av Poincaré-kutt for Hénon-Heiles problemet. I dette tilfellet vil Poincaré-kuttet bli beregnet ved å registrere og plotte $p_2$ mot $q_2$ når $q_1$-komponenten skifter fortegn gitt at $p_1$ er positiv. Dette vil si noe om dynamikken i systemet og vise utviklingen over tid.

### Funksjoner

```python
def poincare_cut(
    y_0,         # (np.ndarray)  - Initial values
    t_0,         # (int)         - Starttime
    T,           # (int)         - Endtime
    h            # (float)       - Step length
):
    # Compute and store points for all methods
    points = {}
    
    # Create processpoolexecutor
    with PPE() as executor:
        # Submit all tasks to PPE "Run all method-functions"
        # Tasks are run on seperate CPUs in parallell
        bogacki_shampine_process = executor.submit(bogacki_shampine, y_0, h, t_0, T, version='Poincare')
        kutta_process = executor.submit(kutta, y_0, h, t_0, T, version='Poincare')
        kahan_process = executor.submit(kahan, y_0, h, t_0, T, version='Poincare')
        stormer_verlet_process = executor.submit(stormer_verlet, y_0, h, t_0, T, version='Poincare')
        
        # Wait for all tasks to complete
        for p in as_completed([bogacki_shampine_process, kutta_process, kahan_process, stormer_verlet_process]): pass
        
        # Retrieve and store results
        points = {
            'Bogacki-Shampine': bogacki_shampine_process.result(),
            'Kutta':            kutta_process.result(),
            'Kahan':            kahan_process.result(),
            'Stormer-Verlet':   stormer_verlet_process.result()
        }
    
    # Create figure
    fig, axes2d = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(9,6), tight_layout=True)
    
    # Plot Poincaré-cuts for all methods
    for idx, item in enumerate(points.items()):
        # Get method and values
        method, val = item
        
        # Fix axes
        ax = axes2d[int(idx > 1) ][int(idx % 2)]
        ax.set_title(method)
        if idx > 1: ax.set_xlabel(r'$q_2$')
        if not idx % 2: ax.set_ylabel(r'$p_2$')
        
        # Plot
        color_change_idx = int(1e4)
        ax.scatter(val[color_change_idx::2], val[color_change_idx + 1::2], s=.8, color='b')
        ax.scatter(val[0:color_change_idx:2], val[1:color_change_idx + 1:2], s=.8, color='orange')
    
    plt.show()
```

### Diskusjon

Derimot når integrasjonlengden øker observeres det at enkelte av metodene slutter å produsere en lukket kurve. Som vi så i oppgave 2 med steglengde $10^{-2}$ (figur 4) drifter Bogacki-Shampine i større grad enn Kutta. Dette kommer også frem i Poincaré-kuttet der Bogacki-Shampine fordreies mer enn Kutta.

Med steglengde $10^{-1}$ resulterer Bogacki-Shampine ikke i en lukket kurve. I tillegg blir Poincaré-kuttet til fordreid. Dette gjenspeiler de store driftene i oppgave 2 (figur 3) med tilsvarende steglengde.

##### Flere invarianter?

En av ideene med Henon-Heiles modelen var å lete etter en ekstra invariant i et system for stjernebevegelse [2, s.15]. Et tegn på at en slik invariant kan eksistere er at poincaré-kuttene blir lukkede kurver. Av våre resultater kan det da tyde på at dette stemmer, men vi vet da fra samme kilde at deler av løsningsrommet er kaotisk slik at denne ideen faller bort.

## 8. Konklusjon

Som plottet for konvergensorden (figur 1) viste ble de fire metodene velykket implementert. 

Deretter ble de energibevarende egenskapene til metodene undersøkt. Konklusjonen er at den beste approksimasjonen for energien over relativt korte tidsperioder er Kuttas metode som skyldes den relativt høye ordenen. For lengre tidsperioder er Störmer-Verlet og Kahan å foretrekke, fordi de har de beste bevaringsegenskapene. 

For korte tidsperioder produserte alle metodene gode poincaré-kutt. Driften i Kutta og Bogacki-Shampine spiller en rolle for lengre tidsperioder og dette kommer tydelig frem i kuttene. I tråd med resultatene i oppgave 2 lykkes Störmer-Verlet også for lange tidsperioder å gjengi gode pointcaré-kutt.
